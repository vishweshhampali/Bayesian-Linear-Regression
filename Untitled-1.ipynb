```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)
X = np.linspace(0, 2*np.pi, 100).reshape(-1, 1)
y = np.sin(X) + np.random.normal(0, 0.1, size=X.shape)

# Visualize original data
plt.figure(figsize=(10, 5))
plt.scatter(X, y, color='blue', label='Original Data')
plt.title('Original Data (Before Transformation)')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
X.shape
```

```python
# Define Gaussian RBF basis function
def gaussian_rbf(x, center, width):
    return np.exp(-0.5 * ((x - center) / width) ** 2)

# Generate centers for Gaussian RBF basis functions
centers = np.linspace(0, 2*np.pi, 10)

# Transform input data using Gaussian RBF basis functions
X_transformed = np.hstack([gaussian_rbf(X, c, 0.1) for c in centers])

# Visualize transformed data
plt.figure(figsize=(10, 5))
plt.plot(X, X_transformed)
plt.title('Data Transformed Using Gaussian RBF Basis Functions')
plt.xlabel('X')
plt.ylabel('Transformed Feature Values')
plt.grid(True)
plt.show()
```
